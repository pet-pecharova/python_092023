```python
# Načtu si data a následně vytvořím stejný regresní model jako v základní části úkolu
import pandas as pd
import statsmodels.formula.api as smf

data = pd.read_csv('df.csv')

formula = 'Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT'
mod = smf.ols(formula=formula, data=data)
res = mod.fit()
res.summary()
```

## Bonus 2
Použij původní (tedy "nerobustní") model a vyčísli Cookovu vzdálenost pro všechny hodnoty. Vyřaď všechny akcie s Cookovou vzdáleností vyšší než 1 a přepočítej regresní model. Nakonec opět najdi potenciálně nejvíce podhodnocenou akcii a podívej se, jestli jde o stejnou akcii jako u předchozích modelů.

```python
# Nechám vyčíslit Cookovu vzdálenost
from statsmodels.stats.outliers_influence import OLSInfluence

influence = OLSInfluence(res)
data['Cook_distance'] = influence.cooks_distance[0]
data.head()
```

```python
# Odstraním všechny hodnoty s Cookovou vzdákeností větší než 1 a sestavím nový model
data = data[data['Cook_distance'] < 1]
mod = smf.ols(formula=formula, data=data)
res = mod.fit()
res.summary()
```

```python
# Provedu odhad a vložím ho jako nový sloupec k původním datům
data['Y_predicted_cook'] = res.predict(data)
data.head()
```

```python
# Do nového sloupce vypočítám rozdíl mezi odhadem a skutečnými cenami
data['Y_difference_cook'] = data['Y_predicted_cook'] - data['Y']
data.head()
```

```python
# Seřadím data podle sloupce s rozdíly v cenách
data.sort_values(by='Y_difference_cook', ascending=False)
```

```python
# Potenciálně nejvíce podhodnocenou akcií je opět akcie č. 79, s rozdílem ceny cca 32.67 (a poté opět akcie č. 6 a nově akcie č. 1).
```
