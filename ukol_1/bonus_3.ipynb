Přidej Sektor (poslední sloupec) jako vysvětlující proměnnou s využitím One Hot Encoding. Podívej se, které sektory mají kladný koeficient a které sektory mají záporný koeficient.

```python
# Načtu si soubor
import pandas as pd
data = pd.read_csv('df.csv')
data.head()
```

```python
# Podívám se na hodnoty ve sloupci 'Sector'
data['Sector'].unique()
```

```python
# Pomocí funkce 'get_dummies' vytvořím novou tabulku, kde je sloupec pro každou z těchto hodnot (True/False)  
sector_gd = pd.get_dummies(data['Sector'])
sector_gd
```

```python
# Původní data a novou tabulku propojím do nové tabulky
data_new = pd.merge(data, sector_gd, left_index=True, right_index=True)
data_new
```

```python
# Některé sloupce (s víceslovným názvem) přejmenuji, abych je mohla uvést v rovnici pro regresní model
data_new.rename(columns={'Financial Services': 'FinancialServices', 'Consumer Defensive': 'ConsumerDefensive',
                         'Consumer Cyclical': 'ConsumerCyclical', 'Communication Services': 'CommunicationServices',
                         'Basic Materials': 'BasicMaterials'}, inplace=True)
data_new.head()
```

```python
# Nechám vytvořit regresní model a zobrazit k němu souhrnnou tabulku
import statsmodels.formula.api as smf

formula = 'Y ~ BasicMaterials + CommunicationServices + Conglomerates + ConsumerCyclical + ConsumerDefensive + Energy + FinancialServices + Healthcare + Industrials + Technology'
mod = smf.ols(formula=formula, data=data_new)
res = mod.fit()
res.summary()
```

```python
# Kladný koeficient mají sektory: Basic Materials, Consumer Cyclical, Consumer Defensive, Financial Services, Technology
# Záporný koeficient mají sektory: Communication Services, Conglomerates, Energy, Healthcare, Industrials
```
