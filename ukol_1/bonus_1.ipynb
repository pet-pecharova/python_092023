## Bonus 1
Sestav model s využitím robustní regrese. Opět proveď vyřazení koeficinetů, které nejsou statisticky významné, a sestav model pouze s významnými koeficienty. Vlož odhady cen do původních dat a opět najdi potenciálně nejvíce podhodnocenou akcii. Jde v případě robustní regese o stejnou akcii, nebo se akcie liší?

```python
# Načtu si data
import pandas as pd
data = pd.read_csv('df.csv')
data.head()
```

```python
# Sestavím model robustní regrese a vypíšu k němu souhrnnou tabulku
import statsmodels.formula.api as smf

formula = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT "
mod = smf.rlm(formula=formula, data=data)
res = mod.fit()
res.summary()
```

```python
# Vyřadím statisticky nevýznamné koeficienty a vytvořím nový model a bez nich
formula = "Y ~ CR + FL + OPM + TAT"
mod = smf.rlm(formula, data=data)
res = mod.fit()
res.summary()
```

```python
# Vytvořím odhad cen a vložím do původní tabulky
data['Y_predicted_robust'] = res.predict(data)
data.head()
```

```python
# Do nového sloupce vypočítám rozdíl mezi odhadem a skutečnou hodnotou
data['Y_difference_robust'] = data['Y_predicted_robust'] - data['Y']
data.head()
```

```python
# Data seřadím podle rozdílu cen
data.sort_values(by='Y_difference_robust', ascending=False)
```

```python
# Potenciálně nejvíce podhodnocenou akcií je opět akcie č. 79, s rozdílem ceny cca 33.28 (a poté opět akcie č. 6 a č. 98).
```
